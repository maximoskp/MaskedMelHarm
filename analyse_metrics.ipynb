```python
import gzip
import json
import numpy as np
 
# Path to your compressed JSON
filepath = 'data/detailed_metrics.gz'
 
# Open and load the nested dict
with gzip.open(filepath, 'rt', encoding='utf-8') as f:
    data = json.load(f)
```

```python
testsets = list( data.keys() )
print(testsets)
models = list( data[testsets[0]].keys() )
print(models)
pieces = {}
pieces[testsets[0]] = list(data[testsets[0]][models[0]].keys())
pieces[testsets[1]] = list(data[testsets[1]][models[0]].keys())
print(pieces)
realgens = list(data[testsets[0]][models[0]][pieces[testsets[0]][0]].keys())
print(realgens)
metrics = list(data[testsets[0]][models[0]][pieces[testsets[0]][0]][realgens[0]].keys())
print(metrics)
```

```python
minval = {m: 100000000 for m in metrics}
maxval = {m: -100000000 for m in metrics}
```

```python
for testset in testsets:
    for model in models:
        for piece in pieces[testset]:
            for metric in metrics:
                try:
                    v = data[testset][model][piece]['real'][metric]
                    if v > maxval[metric]:
                        maxval[metric] = v
                    if v < minval[metric]:
                        minval[metric] = v
                except:
                    print(testset, model, piece, metric)
```

```python
print(minval)
print(maxval)
```

```python
for testset in testsets:
    for model in models:
        for piece in pieces[testset]:
            for metric in metrics:
                try:
                    v = data[testset][model][piece]['gen'][metric]
                    if v > maxval[metric]:
                        maxval[metric] = v
                    if v < minval[metric]:
                        minval[metric] = v
                except:
                    print(testset, model, piece, metric)
```

```python
print(minval)
print(maxval)
```

```python
normdiffs = {k: {d: { m: [] for m in metrics } for d in models} for k in testsets}
print(normdiffs)
```

```python
for testset in testsets:
    for model in models:
        for piece in pieces[testset]:
            for metric in metrics:
                try:
                    v = abs( data[testset][model][piece]['gen'][metric] - data[testset][model][piece]['real'][metric] )
                    v /= maxval[metric] - minval[metric]
                    normdiffs[testset][model][metric].append(v)
                except:
                    print(testset, model, piece, metric)
```

```python
for t in testsets:
    for d in models:
        for m in metrics:
            print(t, d, m, np.mean(normdiffs[t][d][m]))
```

```python
import pandas as pd
```

```python
test_df = pd.DataFrame(np.nan, index=models, columns=metrics)
jazz_df = pd.DataFrame(np.nan, index=models, columns=metrics)
```

```python
for d in models:
    for m in metrics:
        test_df.loc[d, m] = np.mean(normdiffs['test_set'][d][m])
        jazz_df.loc[d, m] = np.mean(normdiffs['jazz_test'][d][m])
```

```python
test_df['mean'] = test_df.mean(axis=1)
jazz_df['mean'] = jazz_df.mean(axis=1)
```

```python
print(test_df)
test_df.to_latex('data/test_df.tex', float_format="%.3f")
```

```python
print(jazz_df)
jazz_df.to_latex('data/jazz_df.tex', float_format="%.3f")
```

```python
from scipy.stats import mannwhitneyu
```

```python
significant = []
for testset in testsets:
    for metric in metrics:
        for model1 in ['base2','base2NPCs','base2NS','random']:
            is_significant = True
            for model2 in ['gpt2','bart']:
                if model1 != model2:
                    set1 = normdiffs[testset][model1][metric]
                    set2 = normdiffs[testset][model2][metric]
                    stat, p_value = mannwhitneyu(set1, set2, alternative='two-sided')
                    if p_value > 0.01:
                        is_significant = False
                        break
            if is_significant:
                significant.append(f'{testset} - {metric} - {model1}')
        for model1 in ['gpt2','bart']:
            is_significant = True
            for model2 in ['base2','base2NPCs','base2NS','random']:
                if model1 != model2:
                    set1 = normdiffs[testset][model1][metric]
                    set2 = normdiffs[testset][model2][metric]
                    stat, p_value = mannwhitneyu(set1, set2, alternative='two-sided')
                    if p_value > 0.01:
                        is_significant = False
                        break
            if is_significant:
                significant.append(f'{testset} - {metric} - {model1}')
```
